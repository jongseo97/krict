```python
from rdkit import Chem
from rdkit.Chem.Crippen import MolLogP
from rdkit import Chem, DataStructs
from rdkit.Chem import AllChem
import numpy as np
import torch
import time
from rdkit.Chem import rdMolDescriptors
from torch.utils.data import Dataset, DataLoader # 이거하면 torch에있는 좋은데이터 사용가능

import torch.nn as nn
import torch.nn.functional as F

#### Preparation of data!!! ####
class DataSet(Dataset):
    
    def __init__(self,smiles):
        morgan_fingerprint = []
        log_p = []
        num_data = len(smiles)
        self.smiles = smiles
        for i in range(num_data):# smiles를 이용해서 분자를 만든다.
            s = smiles[i]
            m = Chem.MolFromSmiles(s)
            # 분자로 fingerprint 받아서(1024차원)
            fp = rdMolDescriptors.GetMorganFingerprintAsBitVect(m,2,nBits=1024)
            logp = MolLogP(m)
            # 두 list에 input x와 label y를 각각 넣는다.
            morgan_fingerprint.append(fp)
            log_p.append(logp)
        #토치화를 시켜서 빠르게할수있게함 이렇게 x랑 y를 만든다.
        #list를 numpy로 바꾸고 그걸 토치텐서로 바꾼거
        self.morgan_fingerprint = torch.from_numpy(np.array(morgan_fingerprint))
        self.log_p = torch.from_numpy(np.array(log_p))

    def __len__(self):#len(a)=3이걸 하려면 이걸 정의해줘야 할수있다.
        return len(self.smiles)

    def __getitem__(self,idx):#getitem쓰고 idx넣었을때 idx에있는 값을 return해라 이렇게 짜면 
    #class data에 몇번째 원소를 뽑아라는게 무난하게 된다. = iterable object로 만드는게 된다.
    #이거 없으면 1000번째있는 logp를 뽑고싶을때 dataset.log_p[1000]이렇게 써야하는데 
    #get item 쓰면 dataset[1000] 이렇게해도 먹힌다.
        sample = dict()
        sample['mfp'] = self.morgan_fingerprint[idx]
        sample['logp'] = self.log_p[idx]
        return sample

#data는 다받았고 이제 모델을 만들자.
class LinearRegressor(nn.Module):
    
    def __init__(self, d_in,d_h,d_out): #init 해야 파라미터가 바뀜
    #15쪽에 박스를 만든다.
        super().__init__()
        self.linear1 = nn.Linear(d_in,d_h)
        self.linear2 = nn.Linear(d_h,d_h)
        self.linear3 = nn.Linear(d_h,d_out)
        self.activation = nn.ReLU() 

    def forward(self, x): #이걸 써서 x를 입력하면 y가 나올수있게한다.
    #15쪽에 박스사이를 연결하는 화살표를 만든다. 진행되도록.
        #relu = self.activation
        retval = self.linear1(x)
        #retval = relu(retval)
        retval = self.linear2(retval)
        #retval = self.activation(retval)
        retval = self.linear3(retval)
        return retval


#data를 로드해주는건데 
####### Loading data ########
def load_data(directory,num):
    f = open(directory,'r')
    smiles_list = []
    for i in range(num):
        line = f.readline()
        words = line.strip().split('\t')
        smiles_list.append(words[-1])
    return smiles_list


# Train: First generate data
smiles_list = load_data('/content/gdrive/My Drive/smiles.txt',10000)
train_data = DataSet(smiles_list[:5000]) # First 5000 molecules = train data
test_data = DataSet(smiles_list[5001:10000]) # Second 5000 molecules = test data

# Prepare data loader
#DataLoader(로드할 데이터, 배치 크기. 트레인은 랜덤성 부여해야해서 셔플해주고 테스트는 어차피 한번씩만 쓰면되니까 셔플안해도됨.
#배치사이즈는 5000개 한번에 넣으면 안되냐? 인데 모델 인풋이 B x 1024인데 배치사이즈가 5000이면 5000 x 1024라서 너무크다.
#그래서 어쩔수없이 배치사이즈를 줄여야한다.
train_data_loader = DataLoader(train_data,batch_size = 128, shuffle = True)
test_data_loader = DataLoader(test_data,batch_size = 128, shuffle = False) # Load data with size 128 (B=128)

# Prepare model
model =  LinearRegressor(1024,256,1) # Construct model


#이제 트레인해야하는데 이건 거의 고정이긴해. 
# Prepare learning parmaters
num_epoch = 10 # Number of scanning whole data 몇번반복해서 쓸거냐
loss_fn = nn.MSELoss() # Mean Square Error Loss: Regression 로스 설정
lr = 1e-4 # Learning rate = 1 x 10^(-4) 뉴럴네트워크 파라미터를 얼마나 크게 바꿀거냐.
optimizer = torch.optim.Adam(model.parameters(),lr=lr) # Adam optimizer 
#옵티마이저는 알고리즘 문젠데 아담 많이쓰더라.  SGD아니면.
#우리는 ㅍ모델의 파라미터를 학습하는거니까 여기에 model.parameters()를 적은거.
# Train process
model.cuda()

for i in range(num_epoch):
    loss_list = []
    for i_batch,batch in enumerate(train_data_loader):
        x_batch = batch['mfp'].float()
        y_batch = batch['logp'].float()#학습할게 스칼라라서 float를 붙였다.
        x_batch = x_batch.cuda()#gpu에 올린다.
        y_batch = y_batch.cuda()
        y_pred = model(x_batch)#예측값을 ypred로 한다.
        y_pred = y_pred.squeeze(1)
        loss = loss_fn(y_pred,y_batch)#로스를계산한다.
        optimizer.zero_grad()
        loss_list.append(loss.data.cpu().numpy())#gpu에있는걸cpu로 내려서 다운받는다.+넘파이가 사용하기 쉬워서 넘파이로.
        loss.backward()#로스를 위에서 어펜드로 저장해놓고 백워드로 로스의 gradient를 계산한다.
        optimizer.step()#optimizer로 loss를 줄이도록 stp()이게 모델에 가하는거라서 모델을 optimize한다.
    avg_loss = np.mean(np.array(loss_list))
    print ('avg loss',avg_loss) #10번(num_epoch) 돌리면서 로스평균을 출력

#model.eval()


######## Validation ########
test_loss_list = []
with torch.no_grad():#이거써야 model에 업데이트가 안된다. 똑같은데 backwardoptimize과정이 없는것.
    for i_batch, batch in enumerate(test_data_loader):
        x_batch = batch['mfp'].float()
        y_batch = batch['logp'].float()
        x_batch = x_batch.cuda()
        y_batch = y_batch.cuda()
        y_pred = model(x_batch)
        y_pred = y_pred.squeeze(1)#dimension문제땜에넣은거.
        loss = loss_fn(y_pred,y_batch)
        loss_list.append(loss.data.cpu().numpy()) 

total_loss = np.mean(np.
                     array(loss_list))
print ('validation loss: ',total_loss)




```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F

from rdkit import Chem
from rdkit.Chem.Crippen import MolLogP

#Dataset
from torch.utils.data import Dataset, DataLoader
from rdkit.Chem.rdmolops import GetAdjacencyMatrix

import numpy as np
import time

######### Prepare data ##########

class GCNDataset(Dataset):

    def __init__(self, max_num_atoms, smiles_list):
        self.max_num_atoms = max_num_atoms
        self.smiles_list = smiles_list
        self.property_list = []
        self.input_feature_list = []
        self.adj_list = []
        self.process_data()
        self.property_list = torch.from_numpy(np.array(self.property_list))

    def process_data(self):
        max_num_atoms = self.max_num_atoms
        for smiles in smiles_list:
            mol = Chem.MolFromSmiles(smiles)
            num_atoms = mol.GetNumAtoms()
            # Get padded adj
            adj = GetAdjacencyMatrix(mol) + np.eye(num_atoms)
            padded_adj = np.zeros((max_num_atoms,max_num_atoms))
            padded_adj[:num_atoms,:num_atoms] = adj
            # Get padded feature list
            padded_feature = np.zeros((max_num_atoms,19))
            feature = []
            for i in range(num_atoms):
                feature.append(self.get_atom_feature(mol,i))
                #print(self.get_atom_feature(mol,i))
            feature = np.array(feature)
            # Get property list
            padded_feature[:num_atoms,:19] = feature
            logp = MolLogP(mol)
            self.property_list.append(logp)
            self.input_feature_list.append(torch.from_numpy(padded_feature))
            self.adj_list.append(torch.from_numpy(padded_adj))

    def __len__(self):
        return len(self.smiles_list)

    def __getitem__(self, idx):
        sample = dict()
        sample['feature'] = self.input_feature_list[idx]
        sample['adj'] = self.adj_list[idx]
        sample['logp'] = self.property_list[idx]
        return sample


    def one_of_k_encoding(self, x, allowable_set):
        """Maps inputs not in the allowable set to the last element."""
        if x not in allowable_set:
            x = allowable_set[-1]
        return list(map(lambda s: x == s, allowable_set))

    def get_atom_feature(self, m, atom_i):

        atom = m.GetAtomWithIdx(atom_i)
        return np.array(self.one_of_k_encoding(atom.GetSymbol(),['C','N','O','F','S','Cl','P','ELSE']) +
                        self.one_of_k_encoding(atom.GetTotalNumHs(), [0, 1, 2, 3, 4, 'ELSE']) +
                        self.one_of_k_encoding(atom.GetFormalCharge(),[-1,0,1,'ELSE']) +
                        [atom.GetIsAromatic()])    # (5, 6, 4, 1) --> total 16


class GConvRegressor(torch.nn.Module):

    def __init__(self, n_feature=128, n_layer = 10):
        super(GConvRegressor, self).__init__()
        layer_list = []
        for i in range(n_layer):
            layer_list.append(nn.Linear(n_feature,n_feature))
        self.W = nn.ModuleList(layer_list)
        self.embedding = nn.Linear(19, n_feature) # Do not use Embedding!
        self.fc = nn.Linear(n_feature, 1)

    def forward(self, x, A):
        x = self.embedding(x)
        for l in self.W:
            x = l(x)
            #x = torch.einsum('ijk,ikl->ijl', (A.clone(), x))
            x = torch.matmul(A,x)
            x = F.relu(x)
        x = x.mean(1)

        retval = self.fc(x)

        return retval

####### Loading data ########
def load_data(directory,maxlen,num):
    f = open(directory,'r')
    smiles_list = []
    cnt = 0
    while cnt < num:
        line = f.readline()
        words = line.strip().split('\t')
        smiles = words[-1]
        if len(smiles) <= maxlen-1:
            cnt += 1
            smiles_list.append(words[-1])
    return smiles_list


# Prepare data
max_natoms = 64
maxlen = 64
smiles_list = load_data('/content/gdrive/My Drive/smiles.txt',maxlen,20000)

# Make train/test/dataloader
train_smiles = smiles_list[:19000]
test_smiles = smiles_list[19000:20000]
train_dataset = GCNDataset(max_natoms,train_smiles)
test_dataset = GCNDataset(max_natoms,test_smiles)

#Dataloader
train_dataloader = DataLoader(train_dataset, batch_size=128, num_workers=1)
test_dataloader = DataLoader(test_dataset, batch_size=128, num_workers=1)

# Make model
model = GConvRegressor(128, 5).cuda()

# Training parameter
lr = 1e-4
optimizer = torch.optim.Adam(model.parameters(), lr=lr)
num_epoch = 20
#optimizer = torch.optim.SGD(model.parameters(), lr=lr)
loss_fn = nn.MSELoss()


loss_list = []
st = time.time()

model.cuda()

for epoch in range(num_epoch):
    loss_list = []
    for i_batch, batch in enumerate(train_dataloader):
        x = batch['feature'].cuda().float()
        y = batch['logp'].cuda().float()
        adj = batch['adj'].cuda().float()
        pred = model(x, adj).squeeze(-1)
        loss = loss_fn(pred, y)
        optimizer.zero_grad()
        torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)
        loss.backward()
        optimizer.step()
        loss_list.append(loss.data.cpu().numpy())
    print (epoch + 1, np.mean(np.array(loss_list)))

end = time.time()
print ('Time:', end-st)
model.eval()
with torch.no_grad():
    loss_list = []
    for i_batch, batch in enumerate(test_dataloader):
        x = batch['feature'].cuda().float()
        y = batch['logp'].cuda().float()
        adj = batch['adj'].cuda().float()
        pred = model(x, adj).squeeze(-1)
        loss = loss_fn(pred, y)
        loss_list.append(loss.data.cpu().numpy())
    print ('validity: ',np.mean(np.array(loss_list)))
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F

from rdkit import Chem
from rdkit.Chem.Crippen import MolLogP

from torch.utils.data import Dataset, DataLoader

import numpy as np
import time

class MolDataset(Dataset):

    def __init__(self,maxlen,c_to_i,smiles_list):
        self.smiles_list = smiles_list
        self.maxlen = maxlen
        self.c_to_i = c_to_i
        sequence_list = self.encode_smiles()
        self.length_list = []
        for sequence in sequence_list:
            self.length_list.append(len(sequence))
        self.sequence_list = sequence_list
        self.length_list = np.array(self.length_list)
        self.length_list = torch.from_numpy(self.length_list)

    def encode_smiles(self):
        smiles_list = self.smiles_list
        c_to_i = self.c_to_i
        maxlen = self.maxlen
        sequence_list = []
        for smiles in smiles_list:
            sequence = []
            for s in smiles:
                sequence.append(c_to_i[s])
            sequence_list.append(torch.from_numpy(np.array(sequence)))
        return sequence_list

    def __len__(self):
        return len(self.smiles_list)

    def __getitem__(self, idx):
        sample = dict()
        sample['length'] = self.length_list[idx]
        sample['seq'] = self.sequence_list[idx]
        return sample

def load_data(directory,maxlen,num):
    f = open(directory,'r')
    smiles_list = []
    cnt = 0
    while cnt < num:
        line = f.readline()
        words = line.strip().split('\t')
        smiles = words[-1]
        if len(smiles) <= maxlen-1:
            cnt += 1
            smiles_list.append(words[-1])
    return smiles_list

def get_c_to_i(smiles_list):
    c_to_i = dict()
    for smiles in smiles_list:
        for letter in smiles:
            if letter not in c_to_i:
                c_to_i[letter] = len(c_to_i)
    return c_to_i

def adjust_smiles(smiles_list,maxlen):
    for i in range(len(smiles_list)):
        smiles_list[i] = smiles_list[i].ljust(maxlen,'X')

def add_X(smiles_list):
    n = len(smiles_list)
    for i in range(n):
        smiles_list[i] += 'X'


#Collate fn
def my_collate(batch):
    sample = dict()
    X =  torch.nn.utils.rnn.pad_sequence([b['seq'] for b in batch], 
                                          batch_first=True, padding_value = 42)
    L = torch.Tensor([b['length'] for b in batch])

    sample['seq'] = X
    sample['length'] = L

    return sample


def print_results(results, msg_length):
    line = []
    for r, ml in zip(results, msg_length):
        r = str(round(r, 5))
        line.append(r.ljust(ml))
    print("\t".join(line))


class RNNRegressor(torch.nn.Module):

    def __init__(self, n_feature=128, n_rnn_layer = 1, n_char=46):
        super().__init__()
        self.n_feature = n_feature
        self.encoder = nn.GRU(input_size=n_feature, hidden_size=n_feature, num_layers=n_rnn_layer)
        self.decoder = nn.GRU(input_size=n_feature*2, hidden_size=n_feature, num_layers=n_rnn_layer)
        
        self.mu = nn.Linear(n_feature, n_feature)
        self.logvar = nn.Linear(n_feature, n_feature)

        self.fc = nn.Linear(n_feature, n_char)
        self.embedding = nn.Embedding(n_char, n_feature)


    def reparameterize(self, mu, logvar):
        std = torch.exp(0.5*logvar)
        eps = torch.randn_like(std) #randn_like: Gaussian distribution sampling
        return mu + eps*std

    def forward(self, x, l):
        x_emb = self.embedding(x) # x_emb: B x L x F
        x_emb = x_emb.permute((1,0,2)) # x_emb: L x B x F
        enc_output, _= self.encoder(x_emb)
        z = []
        for i in range(len(l)):
            z.append(enc_output[l[i]-1,i])
        z = torch.stack(z, 0)

        mu = self.mu(z)
        logvar = self.logvar(z)
        sampled_z = self.reparameterize(mu, logvar) # z: B x F
        dec_input = torch.cat([torch.ones(1, x_emb.size(1), x_emb.size(2)).cuda(), x_emb], 0) # dec_input: (L+1) x B x F: Start codon + initial string
        repeat_z = sampled_z.unsqueeze(0).repeat(dec_input.size(0), 1, 1) # z: (L+1) x B x F
        dec_input = torch.cat([dec_input, repeat_z], -1)
        dec_output, _ = self.decoder(dec_input)
        dec_output = dec_output[:-1]
        dec_output = self.fc(dec_output)

        loss1 = 0
        loss_fn = torch.nn.CrossEntropyLoss(reduction='mean')
        for i in range(len(l)):
            loss1 += loss_fn(dec_output[:(l[i]+1),i,:], x[i,:(l[i]+1)])
        loss2 = -0.5 * torch.sum(1 + logvar - mu.pow(2) - logvar.exp())

        loss1/=len(dec_output)      
        loss2/=len(dec_output)

        return loss1, loss2

    def indices_to_string(self, indices, i_to_c):
        return ''.join([i_to_c[i] for i in indices])

    def sample(self):
        sampled_z = torch.randn(1, self.n_feature).cuda()
        token = torch.ones((1, self.n_feature)).cuda()
        initial_state = torch.zeros((1, 1, self.n_feature)).cuda()

        gen_smiles = []
        maxlen = 64
        for i in range(maxlen):
            dec_input = torch.cat([token, sampled_z], -1).unsqueeze(0)
            output, initial_state = self.decoder(dec_input, initial_state)
            output = F.softmax(self.fc(output), -1).squeeze()
            #if i<1: print (output)
            m = torch.distributions.categorical.Categorical(output)
            sampled_idx = m.sample()
            gen_smiles.append(int(sampled_idx))
            token = self.embedding(torch.tensor([sampled_idx]).cuda())
        return gen_smiles

# Make data
maxlen = 64
smiles_list = load_data('/content/gdrive/My Drive/smiles.txt',maxlen,20000)
c_to_i = get_c_to_i(smiles_list)
add_X(smiles_list)
c_to_i['X'] = len(c_to_i)
i_to_c = dict()
for char in c_to_i:
    index = c_to_i[char]
    i_to_c[index] = char

#Dataset
train_smiles = smiles_list[:10000]
test_smiles = smiles_list[10000:]
train_dataset = MolDataset(maxlen,c_to_i,train_smiles)
test_dataset = MolDataset(maxlen,c_to_i,test_smiles)

train_dataloader = DataLoader(train_dataset, batch_size=128, num_workers=1, collate_fn=my_collate, shuffle=True)
test_dataloader = DataLoader(test_dataset, batch_size=128, num_workers=1, collate_fn=my_collate)

print ('len',len(c_to_i))

#Train model
import time
lr = 1e-4
num_epoch = 51
model = RNNRegressor(512, 1, len(c_to_i)).cuda()
optimizer = torch.optim.Adam(model.parameters(), lr=lr)
loss_fn = nn.MSELoss()
ratio = 10


loss_list = []
st = time.time()

title_list = ['epoch', 'construction loss','kl loss']
length_list = [len(title) for title in title_list]

first_msg = title_list[0] + '\t' + title_list[1] + '\t' + title_list[2]
epoch_list = [5,10,50]


print (first_msg)
for epoch in range(num_epoch):
    epoch_loss = []
    for i_batch, batch in enumerate(train_dataloader):
        model.zero_grad()
        x = batch['seq'].cuda().long()
        l = batch['length'].cuda().long()
        loss1, loss2 = model(x, l)
        loss = loss1 + loss2*ratio
        loss.backward()
        torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)

        optimizer.step()
        epoch_loss.append([loss1.data.cpu().numpy(), loss2.data.cpu().numpy()])
    ####### See reconstruction
    if epoch in epoch_list:
        #fn = f'/content/gdrive/My Drive/Colab Notebooks/save/save_{epoch}.pt'
        fn = f'{epoch}.pt'
        torch.save(model.state_dict(), fn)
    if True: 
        results = [epoch, np.mean(np.array(epoch_loss)[:,0]),np.mean(np.array(epoch_loss)[:,1])]
        print_results(results,length_list)
end = time.time()
print ('Time:', end-st)


'''
######### generate molecules
from rdkit import Chem

model = RNNRegressor(512, 1, len(c_to_i)).cuda()

for epoch in epoch_list:
    #fn = f'/content/gdrive/My Drive/Colab Notebooks/save/save_{i}.pt'
    fn = f'{epoch}.pt'
    model.load_state_dict(torch.load(fn))

    generated = []
    for i in range(100):
        sequence = model.sample()
        smiles = model.indices_to_string(sequence,i_to_c) # Contains X
        smiles = smiles.split('X')[0] # Only gather smiles right before X
        generated.append(smiles)    
    cnt = 0
    well_generated_molecule = []
    for g in generated:
        if Chem.MolFromSmiles(g) is not None: 
            well_generated_molecule.append(g)
            cnt += 1
    for g in well_generated_molecule:
        print (g)
    print (cnt, 'molecules are well generated!')
'''
```
